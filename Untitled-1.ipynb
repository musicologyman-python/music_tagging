```python
from pathlib import Path

MUSIC_DIR = Path('/Users/richard/Music/Music/Media.localized/Music')

mahler_1_directories = [p for p in MUSIC_DIR.iterdir() 
                        if p.name.startswith('Mahler, Symphony No. 1')]

for d in mahler_1_directories:
    print(f'{d.relative_to(MUSIC_DIR)}')
```

```python
bernstein_dir = mahler_1_directories[1]
print(f'{bernstein_dir}')
```

```python
files = sorted(p.stem for p in bernstein_dir.iterdir() if p.suffix == '.m4a')
for f in files:
    print(f)
```

```python
import csv
with Path(MUSIC_DIR / "mahler_symphony_no_1_movements.csv").open(mode='w', encoding='utf-8') as fp:
    writer = csv.DictWriter(fp, fieldnames=['movement_number', 'movement_name'])
    writer.writeheader()
    for i, movement in enumerate(files, start=1):
        writer.writerow( 
        {
            'movement_number': i,
            'movement_name': regex.sub(r'^[IV]+\.\s+', '', movement)
        })
```
